```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report

data = pd.read_csv('/content/movie_prediction.csv')

data = data[:500]
data = data.drop(['Wiki Page'], axis=1)

text_columns = ['Title', 'Origin/Ethnicity', 'Cast', 'Plot']
data['Combined_Text'] = data[text_columns].fillna('').apply(lambda row: ' '.join(map(str, row)), axis=1)

X_text = data['Combined_Text']
y = data['Genre']

X_train, X_test, y_train, y_test = train_test_split(X_text, y, test_size=0.2, random_state=42)

svm_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words='english')),
    ('classifier', SVC(kernel='linear', C=1.0, random_state=42))
])

naive_bayes_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words='english')),
    ('classifier', MultinomialNB())
])

logistic_regression_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words='english')),
    ('classifier', LogisticRegression())
])

# Train and evaluate SVM
svm_pipeline.fit(X_train, y_train)
y_pred_svm = svm_pipeline.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
class_report_svm = classification_report(y_test, y_pred_svm)

# Train and evaluate Naive Bayes
naive_bayes_pipeline.fit(X_train, y_train)
y_pred_nb = naive_bayes_pipeline.predict(X_test)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
class_report_nb = classification_report(y_test, y_pred_nb)

# Train and evaluate Logistic Regression
logistic_regression_pipeline.fit(X_train, y_train)
y_pred_lr = logistic_regression_pipeline.predict(X_test)
accuracy_lr = accuracy_score(y_test, y_pred_lr)
class_report_lr = classification_report(y_test, y_pred_lr)

print('Support Vector Machine (SVM) Classifier:')
print(f'Accuracy: {accuracy_svm:.2f}')
print('Classification Report:\n', class_report_svm)

print('\nNaive Bayes Classifier:')
print(f'Accuracy: {accuracy_nb:.2f}')
print('Classification Report:\n', class_report_nb)

print('\nLogistic Regression Classifier:')
print(f'Accuracy: {accuracy_lr:.2f}')
print('Classification Report:\n', class_report_lr)
```
